# Introduction

Following Carr, et al. ([2019](https://doi.org/10.48550/arXiv.1705.05567)), the dependence of some astrophysical observable $A$ on the PBH abundance or mass function $\phi(M)$ can be expanded as follows:

\begin{equation}
A[\phi(M)] = A_0 + \int\,\mathrm{d}M\,\phi(M)K_1(M) + \int\,\mathrm{d}M_1\mathrm{d}M_2\,\phi(M_1)\phi(M_2)K_2(M_1,M_2) + \cdots,
\end{equation}
where $A_0$ is the background contribution on the observable and $K$ encapsulates all the physics (both intrinsic to the system and relating to detector sensitivity) involved with the system.

If a measurement places an upper bound $A_\mathrm{exp} \geq A[\phi(M)]$ on the observable, then for a monochromatic mass function with $M = M_c$

\begin{equation}
\phi_\mathrm{mon}(M) = f_\mathrm{PBH}(M_c)\delta(M-M_c),
\end{equation}

the upper bound on the observable translates to: 

\begin{equation}
A_\mathrm{exp} - A_0 \geq f_\mathrm{PBH}(M_c)K_1(M_c) + (f_\mathrm{PBH}(M_c))^2K_2(M_c, M_c) + \cdots.
\end{equation}

In the case of observables generated by independent PBHs (i.e. lensing, evaporation, NS capture), $K_n = 0$ for $n \geq 2$, which sets an upper bound on the abundance $f_\mathrm{max}(M_c)$ given by

\begin{equation}
f_\mathrm{PBH}(M_c) \leq \frac{A_\mathrm{exp} - A_0}{K_1(M_c)} = f_\mathrm{max}(M_c).
\end{equation}

In the case of resolvable PBH mergers, two PBHs are always involved in the merge process, which means all $K_n = 0$ except for $K_2$. We can also set $A_0 = 0$ provided we choose a cut-off redshift $z_\mathrm{cut-off}$ large enough to eliminate all potential astrophysical sources of GW merger signals. Under this framework, the expression for the abundance upper bound set by monochromatic mass functions is given by:

\begin{equation}
(f_\mathrm{PBH}(M_c))^{2} \leq \left(\frac{A_\mathrm{exp}}{K_2(M_c, M_c)}\right) = (f_\mathrm{max}(M_c))^2,
\end{equation}

where, following De Luca, et al. ([2021](https://doi.org/10.1088/1475-7516/2021/11/039)), $K_2$ is the redshift-integrated PBH merger event rate per squared-unit mass

\begin{equation}
K_2(M_1, M_2) = \int\,\mathrm{d}z\,\frac{1}{1+z}\frac{\mathrm{d}V_c(z)}{\mathrm{d}z}\frac{\mathrm{d}^2R_\mathrm{PBH}}{\mathrm{d}M_1\mathrm{d}M_2}p_\mathrm{det}(M_1, M_2, z)\frac{1}{\phi(M_1)\phi(M_2)}.
\end{equation}

The constraining procedure for PBH mergers with extended mass functions is thus

\begin{equation}
\int\,\mathrm{d}M_1\mathrm{d}M_2\,\frac{\phi(M_1)\phi(M_2)}{f_\mathrm{max}(M_1)f_\mathrm{max}(M_2)} \leq 1.
\end{equation}

Substituting the above abundance inequality into the constraint expression, we get

\begin{equation}
A_\mathrm{exp} \geq \int\,\mathrm{d}M_1\mathrm{d}M_2\,\phi(M_1)\phi(M_2)K_2(M_1, M_2).
\end{equation}

Note that this is _identical_ to the constraint expression for PBH merger rate, bounded from above by $N_\mathrm{det} = A_\mathrm{exp}$,

\begin{equation}
N_\mathrm{det} \geq \int\,\mathrm{d}z\mathrm{d}M_1\mathrm{d}M_2\,\frac{1}{1+z}\frac{\mathrm{d}V_c(z)}{\mathrm{d}z}\frac{\mathrm{d}^2R_\mathrm{PBH}}{\mathrm{d}M_1\mathrm{d}M_2}p_\mathrm{det}(M_1, M_2, z).
\end{equation}

Thus in order to constrain the abundance, we need to find $f_\mathrm{PBH}$ which satisfies the inequality above. For this, we fix the redshift interval of observation from a $1000 > z > 20$ so as to minimize the possible mergers resulting from ABHs. From here, we select between lognormal and power law distributions, following Carr, et al. ([2019](https://doi.org/10.48550/arXiv.1705.05567)) in their determination of $f_\mathrm{max}$. We set $A_\mathrm{exp} = N_\mathrm{det} = 1$ for all cases, signifying a detection. $\sigma = 0.5$, $M_c$ from 10 to 1000. $M_\mathrm{min} \approx 4\times 10^{14}\,\mathrm{kg}$, $M_\mathrm{max} = 100M_c$

# Merger rate setup

The typical PBH differential merger rate, accounting for suppression factors due to neighboring PBHs and accretion, is given by

\begin{align}
\frac{\mathrm{d}^2R_\mathrm{PBH}}{\mathrm{d}M_1\mathrm{d}M_2} &= \frac{1.6\times 10^6}{\mathrm{Gpc}^3\,\mathrm{yr}}f^{53/37}_\mathrm{PBH, i}\left(\frac{(M_1 + M_2)^2}{M_1M_2}\right)^{34/37}\left(\frac{t}{t_0}\right)^{-34/37}\left(\frac{M^i_1 + M^i_2}{M_\odot}\right)^{-32/37}\times\\
&\times S(M^i_1, M^i_2, f_\mathrm{PBH,i})A(M^i_1, M^i_2, M_1, M_2)\phi(M^i_1)\phi(M^i_2),
\end{align}

where $S$ and $A$ are the suppression factors due to neighboring bodies and accretion, respectively, and the script $i$ denotes values at time of formation. For now, we ignore the contribution of accretion and set $A = 1$. The dominant suppression factor takes the form:

\begin{equation}
S \approx 1.42\left[\frac{\langle m^2\rangle/\langle m\rangle^2}{\bar{N}(y) + C(f_\mathrm{PBH})} + \frac{\sigma^2_M}{f^2_\mathrm{PBH}}\right]e^{-\bar{N}(y)},
\end{equation}

where the rescaled variance of matter perturbations has value $\sigma^2_M \simeq 3.6\times 10^{-5}$, and

\begin{equation}
\bar{N}(y) = \frac{M_\mathrm{tot}}{\langle m\rangle}\frac{f_\mathrm{PBH}}{f_\mathrm{PBH} + \sigma_M}.
\end{equation}

The full expression for $N_\mathrm{det}$ is thus:

\begin{align}
N_\mathrm{det}(f_\mathrm{PBH}, M_c; \sigma) =& \frac{1.6\times 10^6}{\mathrm{Gpc}^3\,\mathrm{yr}} f^{53/37}_\mathrm{PBH} \int\,\mathrm{d}z\, \frac{1}{1+z}\frac{\mathrm{d}V_c(z)}{\mathrm{d}z} \left(\frac{t(z)}{t_0}\right)^{-34/37}\times\\
&\times\int\,\mathrm{d}M_1\,\mathrm{d}M_2\, p_\mathrm{det}(M_1, M_2, z) S(M_1, M_2, f_\mathrm{PBH}) \left(\frac{(M_1 + M_2)^2}{M_1M_2}\right)^{34/37}\left(\frac{M^i_1 + M^i_2}{M_\odot}\right)^{-32/37}\phi(M^i_1; M_c, \sigma)\phi(M^i_2; M_c, \sigma),
\end{align}

where $M_c$ is the reference mass of the distribution and $\sigma$ is the mass distribution width.

We first code for $p_\mathrm{det}$ by considering the function $P(\omega)$

\begin{equation}
p_\mathrm{det} = P(\omega) = a^{(n)}_2[(1-\omega/\alpha^{(n)})^2] + a^{(n)}_4[(1-\omega/\alpha^{(n)})^4] + a^{(n)}_8[(1-\omega/\alpha^{(n)})^8] + (1 - a^{(n)}_2 - a^{(n)}_4 - a^{(n)}_8)[(1-\omega/\alpha^{(n)})^{10}],
\end{equation}

where the coefficients $a^{(n)}_i$ are fit parameters that vary depending on the number of detectors $n$ in the system, and $\alpha^{(n)}$ is the maximum value of $\omega$ for a given number of detectors. Since the LISA observatory is a three-detector system, our coefficients take the values: $a^{(3)}_2 = 1.19549$, $a^{(3)}_4 = 1.61758$, $a^{(3)}_8 = -4.87024$, $\alpha^{(3)} = 1.4$.

```python
def cumulative_P(w):
    a2 = 1.19549
    a4 = 1.61758
    a8 = -4.87024
    alpha = 1.4
    
    signal_ratio = w/alpha
    
    P = [0.0 if signal >= 1.0 else a2*(np.power(1 - signal, 2.0)) + a4*(np.power(1 - signal, 4.)) 
         + a8*(np.power(1 - signal,8.0)) + (1 - a2 - a4 - a8)*(np.power(1 - signal, 10.0)) for signal in signal_ratio]
    
    return P
```

```python
def cumulative_P_alt(w):
    k = 7.
    a = 0.6
    
    return (1.0 + np.exp(k*(w - a)))**(-1.0)
```

Note here that $w = \rho_\mathrm{thr}/\rho_\mathrm{opt}(M_1, M_2, z)$, where $\rho_\mathrm{thr} = 8$ is the minimum SNR usually accepted for a significant signal. In order to compute for $\rho_\mathrm{opt}(M_1, M_2, z)$, we utilize the GW observatory simulation package `gwent`. 

```python
w = np.linspace(0.0, 2.0, 100)

P = cumulative_P(w)
P_alt = cumulative_P_alt(w)

plt.plot(w, P, label='Dominik, et al. 2015')
plt.plot(w, P_alt, label='Fermi-Dirac inspired')
plt.xlabel(r'$w$')
plt.ylabel(r'$P(w)$')
plt.legend()
plt.show()
```

We start by initializing the source binaries that LISA will detect. For the sources, we use the standard non-precessing IMRPhenomD waveforms.

```python
def Initialize_Source(instrument,approximant='pyPhenomD',lalsuite_kwargs={}):
    """Initializes a source binary based on the instrument type and returns the source

    Parameters
    ----------
    instrument : object
        Instance of a gravitational wave detector class
    approximant : str, optional
        the approximant used to calculate the frequency domain waveform of the source.
        Can either be the python implementation of IMRPhenomD ('pyPhenomD', the default) given below,
        or a waveform modelled in LIGO's lalsuite's lalsimulation package.
    lalsuite_kwargs: dict, optional
        More specific user-defined kwargs for the different lalsuite waveforms
    """

    #q = m2/m1 reduced mass
    q = 1.0
    q_min = 1.0
    q_max = 18.0
    q_list = [q,q_min,q_max]

    #Chi = S_i*L/m_i**2, spins of each mass i
    chi1 = 0.0 #spin of m1
    chi2 = 0.0 #spin of m2
    chi_min = -0.85 #Limits of PhenomD for unaligned spins
    chi_max = 0.85
    chi1_list = [chi1,chi_min,chi_max]
    chi2_list = [chi2,chi_min,chi_max]


    if isinstance(instrument,detector.GroundBased):
        #Total source mass
        M_ground_source = [10.,1.,1e4]
        #Redshift
        z_ground_source = [0.1,z0,z_max]

        source = binary.BBHFrequencyDomain(M_ground_source,
                                           q_list,
                                           z_ground_source,
                                           chi1_list,
                                           chi2_list,
                                           approximant=approximant,
                                           lalsuite_kwargs=lalsuite_kwargs)
    elif isinstance(instrument,detector.SpaceBased):
        M_space_source = [1e6,1.,1e10]
        z_space_source = [1.0,z0,z_max]
        source = binary.BBHFrequencyDomain(M_space_source,
                                           q_list,
                                           z_space_source,
                                           chi1_list,
                                           chi2_list,
                                           approximant=approximant,
                                           lalsuite_kwargs=lalsuite_kwargs)
    elif isinstance(instrument,detector.PTA):
        M_pta_source = [1e9,1e8,1e11]
        z_pta_source = [0.1,z0,z_max]
        source = binary.BBHFrequencyDomain(M_pta_source,
                                           q_list,
                                           z_pta_source,
                                           chi1_list,
                                           chi2_list,
                                           approximant=approximant,
                                           lalsuite_kwargs=lalsuite_kwargs)
    return source
```

Next, we initialize our detector parameters. For now, we stick to LISA.

```python
def Initialize_LISA():
    #Values taken from the ESA L3 proposal, Amaro-Seaone, et al., 2017 (https://arxiv.org/abs/1702.00786)
    T_obs_space_list = [4*u.yr,1*u.yr,10*u.yr]

    #armlength in meters
    L = 2.5e9*u.m
    L_min = 1.0e7*u.m
    L_max = 1.0e11*u.m
    L_list = [L,L_min,L_max]

    #Acceleration Noise Amplitude
    A_acc = 3e-15*u.m/u.s/u.s
    A_acc_min = 1e-16*u.m/u.s/u.s
    A_acc_max = 1e-14*u.m/u.s/u.s
    A_acc_list = [A_acc,A_acc_min,A_acc_max]

    #The Low Acceleration Noise Break Frequency
    f_acc_break_low = .4*u.mHz.to('Hz')*u.Hz
    f_acc_break_low_min = .1*u.mHz.to('Hz')*u.Hz
    f_acc_break_low_max = 1.0*u.mHz.to('Hz')*u.Hz
    f_acc_break_low_list = [f_acc_break_low,f_acc_break_low_min,f_acc_break_low_max]

    #The High Acceleration Noise Break Frequency
    f_acc_break_high = 8.*u.mHz.to('Hz')*u.Hz
    f_acc_break_high_min = 1.*u.mHz.to('Hz')*u.Hz
    f_acc_break_high_max = 10.*u.mHz.to('Hz')*u.Hz
    f_acc_break_high_list = [f_acc_break_high,f_acc_break_high_min,f_acc_break_high_max]

    #The Optical Metrology Noise Break Frequency
    f_IFO_break = 2.*u.mHz.to('Hz')*u.Hz
    f_IFO_break_min = 1.*u.mHz.to('Hz')*u.Hz
    f_IFO_break_max = 10.*u.mHz.to('Hz')*u.Hz
    f_IFO_break_list = [f_IFO_break,f_IFO_break_min,f_IFO_break_max]

    #Detector Optical Metrology Noise
    A_IFO = 10e-12*u.m
    A_IFO_min = 1.0e-13*u.m
    A_IFO_max = 1.0e-10*u.m
    A_IFO_list = [A_IFO,A_IFO_min,A_IFO_max]

    #Unresolved Galactic WD Background
    Background = False

    #Numerical Transfer Function
    T_type = 'N'

    LISA_prop1 = detector.SpaceBased('LISA_prop1',
                                     T_obs_space_list,L_list,A_acc_list,
                                     f_acc_break_low_list,f_acc_break_high_list,
                                     A_IFO_list,f_IFO_break_list,
                                     Background=Background,T_type=T_type)
    return LISA_prop1
```

The result we wish to obtain from this setup is the SNR as a function of mass and redshift. For this, we will initialize a sample rate of $100$ for both $M$ and $z$ to obtain a $100 \times 100$ SNR matrix.

```python
#samples along redshift
sampleRate_z = 100
#samples along mass
sampleRate_M = 200
```

Now we calculate for the LISA SNR given these ranges.

```python
#Variable on y-axis
var_y = 'z'
#Variable on x-axis
var_x = 'M'

instrument = Initialize_LISA()
source = Initialize_Source(instrument)
start = time.time()
[sample_x,sample_y,SNRMatrix] = snr.Get_SNR_Matrix(source,instrument,
                                                   var_x,sampleRate_M,
                                                   var_y,sampleRate_z)
end = time.time()

print('Model: ',instrument.name + '_' + var_x + '_vs_' + var_y,',',' done. t = : ',end-start)
Plot_SNR(var_x,sample_x,var_y,sample_y,SNRMatrix,smooth_contours=False)
```

```python
from scipy import interpolate

SNR_interp = interpolate.interp2d(sample_x, sample_y, SNRMatrix, kind='cubic')
```

```python
#from joblib import Parallel, delayed

#def loguniform(low, high, size=None, base=10.0):
#    exp_low = np.log10(low)
#    exp_high = np.log10(high)
#    return np.power(base, np.random.uniform(exp_low, exp_high, size))

def MC_loginteg_triple(func, M_c, var, z0, z1, n=100):
    
    x_list = np.random.lognormal(np.log(M_c), var, n)
    y_list = np.random.lognormal(np.log(M_c), var, n)
    z_list = np.linspace(z0, z1, n)
    
    X, Y, Z = np.meshgrid(x_list, y_list, z_list, indexing='ij', sparse=True)
    
    #f = Parallel(n_jobs=4)(delayed(func)(x, y, z) for z in z_list for y in y_list for x in x_list)
    #f = [func(x,y,z) for z in z_list for y in y_list for x in x_list]
    #f_comp = np.fromiter(f, 'float')
    f_comp = func(X, Y, Z)

    f_mean = f_comp.sum()/(len(f_comp)**3.0)
    value = (x_list.max() - x_list.min())*(y_list.max() - y_list.min())*(z1 - z0)*f_mean

    return value
```

```python
def logdist(M, M_c, sigma):
    norm = np.power(np.sqrt(2*np.pi) * sigma * M, -1.0)
    
    return norm * np.exp(-0.5*np.power(np.log(M/M_c)/sigma, 2.0))
```

```python
def nbar(M1, M2, M_ave, f):
    M_tot = M1 + M2
    
    return (M_tot/M_ave)*(f/(f + np.sqrt(sigmaM_squared)))
```

```python
import scipy.special as sc

def cfit(M_ave, M_var, f):
    fs = np.power(f, 2.0)/sigmaM_squared
    hyp = (0.255*sc.hyperu(21.0/74.0, 0.5, (5.0/6.0)*fs))**(-74.0/21.0)
    
    return (M_var/np.power(M_ave, 2.0))*fs/(hyp - 1.0)
```

```python
def supp(M1, M2, M_ave, M_var, f):
    factor = np.power((M_var/np.power(M_ave, 2.0))/(nbar(M1, M2, M_ave, f) + cfit(M_ave, M_var, f)) 
                    + sigmaM_squared/np.power(f,2.0), -21.0/74.0)
    
    return 1.42*factor*np.exp(-nbar(M1, M2, M_ave, f))
```

```python
def Mzdep(M1, M2, z):
    dV = cosmo.differential_comoving_volume(z)
    t = cosmo.age(z)
    a = cosmo.scale_factor(z)
    
    M_tot = M1 + M2
    M_square = M1*M2
    
    M_factor = np.power(np.power(M_tot, 2.0)/M_square, 34.0/37.0)
    M_dep = np.power(M_tot, -32.0/37.0)
    
    rho_thr = 15.0
    rho_opt = np.sqrt(SNR_interp(M_tot.flatten(), z.flatten()))
    p_det = cumulative_P_alt(rho_thr/rho_opt.flatten())[0]
    
    return a*(dV.value)*p_det*(np.power(t.value/t0, -34.0/37.0))*M_factor*M_dep
```

```python
def Ndet(M_ave, M_var, M_c, sigma, f):
    integ = lambda M1, M2, z: Mzdep(M1, M2, z)*supp(M1, M2, M_ave, M_var, f)
    
    f_eval = np.power(f, 53.0/37.0)
    Mz_eval = MC_loginteg_triple(integ, M_c, sigma, z0, 1.0, 200)
    
    return (1.6*1e6) * f_eval * Mz_eval
```

We consider the mass range from $M_\mathrm{min} \sim 1\,M_\odot$ to $M_\mathrm{max} \approx 10^{10}\,M_\odot$ and the PBH abundance range $10^{-20} < f_\mathrm{PBH} < 10^0$.

```python
sig = 2.0
#mass_max_array = sample_x*np.power(10.0, 2.0*sig)
#mass_min_array = sample_x*np.power(10.0, -2.0*sig)

mass_ave = sample_x*np.exp(sig**2.0 / 2.0)
mass_var = np.power(sample_x*np.exp(sig**2.0), 2.0)
```

```python
from IPython.display import display, clear_output

fPBH_samples = np.logspace(-20.0, 0.0, sample_x.size)
fPBH_array = []
mass_array = []

start_time = time.time()
iteration = 0

for mass, ave, var in zip(sample_x, mass_ave, mass_var):
    clear_output(wait=True)
    iteration += 1
    display("Progress: "+str(iteration)+"/"+str(sample_x.size))
    time.sleep(1)
    for fPBH in fPBH_samples:
        N = Ndet(ave, var, mass, sig, fPBH)
        if N >= 1.0:
            fPBH_array.append(fPBH)
            mass_array.append(mass)
            break

end_time = time.time()

print("Total runtime: ", end_time - start_time)
```
